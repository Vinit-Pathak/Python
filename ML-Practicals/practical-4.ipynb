```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mtp
```

```python
dataset = pd.read_csv('E:\Python\ML-Practicals\Datasets\Level.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,1:2].values
x
```

```python
y = dataset.iloc[:,2].values
y
```

```python
from sklearn.linear_model import LinearRegression
lin_regs = LinearRegression()
lin_regs.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regs = PolynomialFeatures(degree=5)
x_poly = poly_regs.fit_transform(x)
lin_regs_2 = LinearRegression()
lin_regs_2.fit(x_poly,y)
```

```python
mtp.scatter(x,y,color="red")
mtp.plot(x,lin_regs.predict(x),color="blue")
mtp.title("Bluff Detection Model(Linear Regression)")
mtp.xlabel("Position Levels")
mtp.ylabel("Salary")
mtp.show()
```

```python
mtp.scatter(x,y,color="red")
mtp.plot(x,lin_regs_2.predict(poly_regs.fit_transform(x)),color="blue")
mtp.title("Bluff Detection Model(Polynomial Regression)")
mtp.xlabel("Position Levels")
mtp.ylabel("Salary")
mtp.show()
```
